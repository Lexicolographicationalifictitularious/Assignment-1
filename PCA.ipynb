```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc                     # for loading image
np.random.seed(1)

# filename structure
path = 'C:/Users/User/Downloads/YALE/unpadded/' # path to the database
ids = range(1,2)
states = ['centerlight', 'glasses', 'happy', 'leftlight',
          'noglasses', 'normal', 'rightlight','sad',
          'sleepy', 'surprised', 'wink' ]
prefix = 'subject'
surfix = '.pgm'

# data dimension
h = 116 # hight
w = 98 # width
D = h * w
N = len(states)*15
K = 1

# collect all data
X = np.zeros((D, N))
cnt = 0
for person_id in range(1,2):
    for state in states:
        fn = path + prefix + str(person_id).zfill(2) + '.' + state + surfix
        X[:, cnt] = misc.imread(fn).reshape(D)
        cnt += 1

# Doing PCA, note that each row is a datapoint
from sklearn.decomposition import PCA
while K <= 51:    
    pca = PCA(n_components=K)
    pca.fit(X.T) # must-ask

    # projection matrix
    U = pca.components_.T

    for person_id in range(1,2):
        for state in ['centerlight']:
            fn = path + prefix + str(person_id).zfill(2) + '.' + state + surfix # concatenate names
            im = misc.imread(fn) # read fn
            plt.axis('off')
    #         plt.imshow(im, interpolation='nearest' )
            f1 = plt.imshow(im, interpolation='nearest')    # show pictures on a graph
            f1.axes.get_xaxis().set_visible(False) # turn off axes in plt.show()
            f1.axes.get_yaxis().set_visible(False)
            # change graph's color to gray
            plt.gray()
            fn = 'ori' + str(person_id).zfill(2) + '.png'
            plt.savefig(fn, bbox_inches='tight', pad_inches=0) # save pictures to memory
            plt.show()
            # reshape and subtract mean, don't forget 
            x = im.reshape(D, 1) - pca.mean_.reshape(D, 1)  # must-ask
            # encode
            z = U.T.dot(x) # compress image
            print(z)
            #decode 
            x_tilde = U.dot(z) + pca.mean_.reshape(D, 1) # restore image
            print(x_tilde)
            # reshape to orginal dim
            im_tilde = x_tilde.reshape(116, 98)
            plt.axis('off')
    #         plt.imshow(im_tilde, interpolation='nearest' )
            f1 = plt.imshow(im_tilde, interpolation='nearest')
            f1.axes.get_xaxis().set_visible(False)
            f1.axes.get_yaxis().set_visible(False)
            plt.gray()
            fn = 'res' + str(person_id).zfill(2) + '.png'
            plt.savefig(fn, bbox_inches='tight', pad_inches=0)
            plt.show()
    K+=10
```
